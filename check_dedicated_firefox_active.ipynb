```python
# check_dedicated_firefox_active()


import psutil

def get_firefox_processes(target_path):
    firefox_processes = []
    try:
        for proc in psutil.process_iter(['pid', 'name', 'exe']):
            if 'firefox' in proc.info['name'].lower() and target_path in proc.info['exe']:
                firefox_processes.append({
                    'pid': proc.info['pid'],
                    'exe': proc.info['exe']
                })
    except Exception as e:
        print(f"Error: {e}")

    return firefox_processes

# Example usage:
target_path = r"C:\Program Files (x86)\Mozilla Firefox\firefox.exe"
firefox_processes = get_firefox_processes(target_path)
for process in firefox_processes:
    print(f"PID: {process['pid']}, Executable Path: {process['exe']}")


print(f"PID: {process['pid']}, Executable Path: {process['exe']}")
print(process['pid'])


# ------------------------- ---------------- ---------------------

import win32gui
import win32process
import time



def get_active_firefox_pid():
    try:
        foreground_window = win32gui.GetForegroundWindow()
        window_title = win32gui.GetWindowText(foreground_window)
        if 'firefox' in window_title.lower():
            _, pid = win32process.GetWindowThreadProcessId(foreground_window)
            return pid
    except Exception as e:
        print(f"Error: {e}")
    return None

if __name__ == "__main__":
    time.sleep(3)
    active_firefox_pid = get_active_firefox_pid()
    if active_firefox_pid:
        print("Active Firefox window PID:", active_firefox_pid)
    else:
        print("No active Firefox window found.")
print(active_firefox_pid)
# ---------- -------------- ------------ ------------ --------------------
# ---------- -------------- ------------ ------------ --------------------
def check_dedicated_firefox_active():
    if process['pid'] == active_firefox_pid:
        print ("Yes")
    else:
        print("No")


check_dedicated_firefox_active()
```
